```python
from wordcloud import WordCloud
import imageio
import matplotlib.pyplot as plt
import jieba
```

```python
def read_data():
    filename = "data.csv"
    file = open(filename).read()
    
    remove = [",","，","\n","。","\t","\s","\xa0", "\0x91", ";","；","：","?","【","】","  ","）","（","、","·","…","/"]
    for i in remove:
        file = file.replace(i, "")
    for i in range(10):
        file = file.replace(str(i), "")
    word = jieba.lcut(file)
    
    return word
    '''
    print(word)
    with open("new_data.txt",'w') as f:
        for i in word:
            f.write(str(i))
    print("job done")
    '''
```

```python
def get_keyword(words):
    count = {}
    
    for word in words:
        count[word] = count.get(word,0) + 1
    
    exclude = ["的","you"," ","in","with","对","与","以上","月",".","等","有","及","并","-","能","你","完成","公司","为","支持","具备","相关",
               "参与","and","使用","to","和",]
    for key in list(count.keys()):
        if key in exclude:
            del count[key]
    
    list_1 = list(count.items())
    list_1.sort(key = lambda x: x[1], reverse = True)
    
    print(list_1[:200])
    with open("new_data.txt",'w') as f:
        for i in list_1[:200]:
            f.write(str(i))
    print("job done")
```

```python
def img_grearte():
    with open("new_data.txt","r") as file:   
        txt=file.read()
    word = WordCloud(font_path = 'simfang.ttf', width = 3000, height = 3000).generate(txt)
    word = WordCloud(width = 3000, height = 3000).generate(txt)
    word.to_file('test_1.png')
    
    plt.imshow(word)    #使用plt库显示图片
    plt.axis("off")
    plt.show()
```

```python
word = read_data()
get_keyword(word)
img_grearte()
```
