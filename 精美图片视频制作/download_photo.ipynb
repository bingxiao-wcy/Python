```python
import requests
import os
import shutil

API_KEY = '50631869-72e105bbffeadd92bbfd65e8f'
SEARCH_URL = 'https://pixabay.com/api/'
VIDEO_SEARCH_URL = 'https://pixabay.com/api/videos/'

# 创建文件夹保存下载的图片和视频
if not os.path.exists('downloaded_images'):
    os.makedirs('downloaded_images')

def download_image(keyword, min_width=1920, min_height=1080, num_images=5):
    # 从 Pixabay 下载图片的函数
    # 参数说明：
    # keyword: 搜索提示词，用于搜索图片
    # min_width: 图片的最小宽度（像素）
    # min_height: 图片的最小高度（像素）
    # num_images: 每页返回的图片数量
    params = {
        'key': API_KEY,  # API密钥，用于身份验证
        'q': keyword,    # 搜索关于图片的关键词，URL编码
        # 'image_type': 'photo',  # 图片类型：photo、illustration、vector等
        'per_page': num_images,  # 每页返回的结果数，默认20，范围3-200
        # 'safesearch': 'true',  # 是否只返回适合所有年龄段的图片，默认false
        'min_width': min_width,  # 图片的最小宽度
        'min_height': min_height,  # 图片的最小高度
        'order': 'latest',  # 按最新排序
        'content_type':'ai'
        # 'orientation': 'vertical' # 竖屏
    }

    response = requests.get(SEARCH_URL, params=params)
    data = response.json()

    if data.get('total') > 0:
        for i, hit in enumerate(data['hits']):
            image_url = hit['largeImageURL']
            image_name = f"{keyword.replace(' ', '_')}_image_{i}.jpg"
            image_path = os.path.join('downloaded_images', image_name)
            
            with open(image_path, 'wb') as f:
                image_data = requests.get(image_url).content
                f.write(image_data)
            print(f"Downloaded image {image_name} with resolution {min_width}x{min_height}")
    else:
        print("No images found")


def move_images_and_rename():
    source_dir = 'downloaded_images'
    target_dir = 'download'
    
    # 创建目标文件夹
    os.makedirs(target_dir, exist_ok=True)
    
    # 获取源文件夹中的所有文件
    source_files = os.listdir(source_dir)
    
    # 遍历每个文件
    for filename in source_files:
        # 构造源文件路径
        source_path = os.path.join(source_dir, filename)
        
        # 构造初始的目标文件路径
        target_path = os.path.join(target_dir, filename)
        
        # 如果目标文件已存在，则在文件名后加上序号，直到找到一个不存在的文件名
        if os.path.exists(target_path):
            base_name, ext = os.path.splitext(filename)
            counter = 1
            while True:
                new_target_path = os.path.join(target_dir, f"{base_name}_{counter}{ext}")
                if not os.path.exists(new_target_path):
                    target_path = new_target_path
                    break
                counter += 1

# 示例使用
download_image('金', min_width=1280, min_height=1280, num_images=200)
move_images_and_rename()
```
