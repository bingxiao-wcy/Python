```python
from PIL import Image, ImageDraw, ImageFont
import time
import pandas as pd
import os

# def addAlpha(img_path,out_img_path,alpha):
#     bg_img =Image.open(img_path).convert('RGBA')

#     width = bg_img.size[0]
#     height = bg_img.size[1]

#     def drawRect(img, pos, **kwargs):
#         transp = Image.new('RGBA', img.size, (0,0,0,0))
#         draw = ImageDraw.Draw(transp, "RGBA")
#         draw.rectangle(pos, **kwargs)
#         img.paste(Image.alpha_composite(img, transp))

#     drawRect(bg_img, (0,0,width,height), fill=(255,255,255,alpha))
#     bg_img.convert('RGB')
#     bg_img.save(out_img_path)

def image_add_text(img_path, text, output_path= 'result\output_text.jpg', alpha=200, flag = 0):
    img = Image.open(img_path).convert('RGBA')
    width = img.size[0]
    height = img.size[1]
    font = 'STXINWEI.TTF'
    text_color = (0,0,0)
    top = 0
    # 计算字体大小

    text_size = int(width/(len(max(text, key=len))+4))
    line = text_size + int(text_size / 2)

    # print(width, len(max(text, key=len))+4, )
    transp = Image.new('RGBA', img.size, (0,0,0,0))
    draw = ImageDraw.Draw(transp, "RGBA")
    fontStyle = ImageFont.truetype(font, text_size, encoding="utf-8")
    x = 0
    y = 0
    
    if flag == 0:  
        #求蒙版的长和宽
        transp_width = fontStyle.getbbox(max(text, key=len) + '宽度')[2] 
        transp_height = len(text) * line + 2.5 * line
        x = ( width - transp_width ) / 2
        y = abs(( height - transp_height ) / 2 )
        # print(max(text, key=len), width, height, x, y,transp_width+x, transp_height+y)
        draw.rectangle((x,y,transp_width+x,transp_height+y),fill=(255,255,255,alpha))
        img.paste(Image.alpha_composite(img, transp))   
    else:
         draw.rectangle((0,0,width,height),fill=(255,255,255,alpha))
         img.paste(Image.alpha_composite(img, transp))
    img.convert('RGB')
    draw = ImageDraw.Draw(img)
    for t in ['每天学习一点点']:
        fontStyle_title = ImageFont.truetype(font, text_size, encoding="utf-8")
        w = fontStyle.getbbox(t)[2]
        draw.text(((width-w)/2, y + line), t, (255,0,0), font=fontStyle_title)
        top = y + 2 * line
    for t in text[:-1]:
        w = fontStyle.getbbox(t)[2]
        draw.text(((width-w)/2, top), t, text_color, font=fontStyle)
        top = top + line
    if text[-1] != '《nan》 nan' and text[-1] != '《 》':
        fontStyle_author = ImageFont.truetype('STXINGKA.TTF', int(text_size / 1.2), encoding="utf-8")
        w = fontStyle_author.getbbox(text[-1])[2]
        draw.text(((width-w)/2, top + int(text_size / 2.4) ), text[-1], (255,0,0), font=fontStyle_author)
        
    img.save(output_path)

text = pd.read_excel('txt/1.xlsx',sheet_name='正在用')
textl = []

for i,row in text.iterrows():
    text_tmp = row[0].replace('，',',').replace('。',',').replace('；',',').replace(';',',').replace('：',',').split(',')
    text_tmp = [i for i in text_tmp if i!='']
    author_tmp = '' + '《' + str(row[1]) + '》 ' + str(row[2])
    text_tmp.append(author_tmp)
    textl.append(text_tmp)
file = os.listdir('img')
file.sort(key=lambda x:int(x[:-4]))
# 102, 102, 0 深棕
# 51, 51, 0 深绿
# 204, 102, 0 橙色

# style = [
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 1     
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 2    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 3    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 4    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 5    
#     [0, 300, (0, 0, 0), 250, 200, 'STXINWEI.TTF', 0],    # 6    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],      # 7    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 8    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0],     # 9    
#     [0, 300, (0, 0, 0), 150, 200, 'STXINWEI.TTF', 0]    # 10    
# ]

style = [
    [200,0],  # 1     
    [200,0],  # 2    
    [200,0],  # 3    
    [200,0],  # 4    
    [200,0],  # 5    
    [200,0],  # 6    
    [200,0],  # 7    
    [200,0],  # 8    
    [200,0],  # 9    
    [200,0],  # 10
    [200,0],  # 11
    [200,0],  # 12
    [200,0]   # 13
]

for i in range(len(textl)):
    print(file[i],style[i],textl[i])
    alpha,flag = style[i][0],style[i][1]
    img_path = 'img\\' + file[i] 
    out_img_path = 'output.png'
    output_path = 'result\output' + '_'  + str(time.time()) + '_' + str(i+1) + '.png'
    # addAlpha(img_path, out_img_path, alpha)
    image_add_text(img_path, textl[i], output_path=output_path, alpha=alpha, flag=flag)

```
